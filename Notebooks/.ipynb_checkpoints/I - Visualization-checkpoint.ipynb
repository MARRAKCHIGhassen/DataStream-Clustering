```python
# Subplotting
def subplot(axe, data_frame, attributes, dimensions, classes) :
    if dimensions == "2D" :
        for classe in range(len(classes)) :
            axe.scatter(data_frame[attributes[0]][data_frame['class'] == classes[classe]], 
                        data_frame[attributes[1]][data_frame['class'] == classes[classe]], 
                        label=classes[classe])
            # Labels
            axe.set_xlabel(attributes[0])
            axe.set_ylabel(attributes[1])

            # Title
            axe.set_title('Visualisation de ' + attributes[0] + attributes[1])
    else :
        # Plotting classes
        for classe in range(len(classes)) :
            axe.scatter3D(data_frame[attributes[0]][data_frame['class'] == classes[classe]], 
                          data_frame[attributes[1]][data_frame['class'] == classes[classe]],
                          data_frame[attributes[2]][data_frame['class'] == classes[classe]],
                          label=classes[classe])
        
        # Labels
        axe.set_xlabel(attributes[0])
        axe.set_ylabel(attributes[1])
        axe.set_zlabel(attributes[2])
        
        # Title
        axe.set_title('Visualisation de ' + attributes[0] + attributes[1] + attributes[2])
        
```

```python
# Combinations
def get_attributes_combinations(columns, dimensions) :
    combinations = dict()
    
    # Construction  
    # All combinations
    if dimensions == '2D' :
        inter_combinations = list(itertools.combinations(columns, 2))
    else :
        inter_combinations = list(itertools.combinations(columns, 3))
    
    # Dimensions
    combinations['dimension'] = dimensions
    combinations['nb_lignes'] = len(columns)
    
    if dimensions == '2D' :
        combinations['nb_colonnes'] = len(columns)
    else :
        combinations['nb_colonnes'] = len(columns)**2
    
    
    # All positions
    lignes = range(1, combinations['nb_lignes'] + 1)
    colonnes = range(1, combinations['nb_colonnes'] + 1)
    
    # Verification of logic
    assert (len(lignes) * len(colonnes)) >= len(inter_combinations)
    print("Enough positions for combinations Verified ! ")
    
    # Inserting {'combis' : list()}
    combinations['combis'] = list()
    
    ligne_combination = 0
    colonne_combination = 0
    for _ in range(len(inter_combinations)) :
        # print(ligne_combination, colonne_combination)
        # Appending {'ligne' : -, 'colonne' : -, 'attributes' : tuple()}
        combi = dict()
        
        combi['ligne'] = lignes[ligne_combination]
        combi['colonne'] = colonnes[colonne_combination]
        
        combi['attributes'] = inter_combinations[_]
        
        combi['position'] = (ligne_combination + 1) * (colonne_combination + 1)
        
        combinations['combis'].append(combi)
        
        # Updating position
        if colonne_combination == (combinations['nb_colonnes'] - ligne_combination - 1) :
            colonne_combination = 0
            ligne_combination += 1
        else :
            colonne_combination += 1
        
    
    
    # Total number of combinations
    combinations['number'] = len(inter_combinations)
    
    
    return combinations
```
